## 4. Date Format Standardization

Converting Order Date and Ship Date columns to datetime format for accurate lead time calculation.

```python
# Convert date columns to datetime format for accurate time calculations

df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d-%m-%Y')
```

```python
df.info()
```

## 5. Lead Time Calculation

Creating shipping lead time metric (Ship Date - Order Date).
This metric will serve as the foundation for route efficiency analysis.

```python
# Calculate shipping lead time in days (core efficiency metric)

df['lead_time'] = (df['Ship Date'] - df['Order Date']).dt.days
```

```python
# Examine distribution of lead time to detect unrealistic shipping durations

df['lead_time'].describe()
```
