# Logistics Key Performance Indicators (KPI) Analysis

## Objective

This notebook calculates key logistics performance indicators to evaluate the efficiency of the shipping network.

These KPIs help quantify delivery performance, shipment demand, and route efficiency, enabling logistics managers to identify operational improvement opportunities.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/processed/route_efficiency_data.csv")

df.head()
```

## KPI 1 — Shipping Lead Time

Shipping Lead Time measures the duration between the shipment date and delivery date.

This metric helps evaluate the speed of the logistics process.

```python
df['Shipping_Lead_Time'] = df['lead_time']
```

## KPI 2 — Average Lead Time per Route

Average lead time represents the mean delivery duration for each shipping route.

```python
avg_route_lead_time = df.groupby("Route")["lead_time"].mean().reset_index()

avg_route_lead_time.head()
```

## KPI 3 — Route Volume

Route volume measures the number of shipments processed on each route.

```python
route_volume = df.groupby("Route").size().reset_index(name="Shipment_Count")

route_volume.head()
```

## KPI 4 — Delay Frequency

Delay frequency measures the percentage of shipments exceeding an acceptable lead-time threshold.

```python
delay_threshold = df['lead_time'].mean()

df['Delayed'] = df['lead_time'] > delay_threshold

delay_frequency = df['Delayed'].mean() * 100

print(round(delay_frequency, 2))
```

## KPI 5 — Route Efficiency Score

The route efficiency score evaluates route performance based on normalized lead-time efficiency.

```python
df['Route_Efficiency_Score'] = (
    df['lead_time'].max() - df['lead_time']
) / df['lead_time'].max()
```

## KPI Insights

The calculated KPIs provide a quantitative overview of the logistics network performance.

Key insights derived from the analysis include:

- Identification of high-volume shipping routes
- Detection of routes with longer average delivery times
- Measurement of overall delay frequency
- Evaluation of route efficiency scores

These insights help logistics managers prioritize operational improvements.

## Business Implications

The KPI analysis provides actionable insights for logistics optimization.

Potential actions include:

- Optimizing routes with high lead times
- Monitoring high-volume shipment routes
- Reducing delay frequency through improved scheduling
- Prioritizing efficient shipping modes

These improvements can enhance delivery reliability and customer satisfaction.
